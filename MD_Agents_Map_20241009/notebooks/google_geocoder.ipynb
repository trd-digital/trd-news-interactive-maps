```python
import pandas as pd
import requests
from os import path
```

```python
DATA_DIR = '/Users/trdny_josephjungermann/Documents/the_real_deal/20241009_MD_Agents_Map/md_agents_map/data'
md_addresses = pd.read_csv(path.join(DATA_DIR, 'md_addresses.csv'))
md_addresses
```

```python
# my google maps geocoding API key
# go to your google cloud profile and "Geocode Mapping" project to retrieve API key and then paste it here
api_key = ''
```

```python
address_list = md_addresses['Address_2'].tolist()
address_list
```

```python
# Function to get latitude and longitude
def get_lat_lng(address):
    url = f'https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={api_key}'
    response = requests.get(url)
    data = response.json()
    
    if data['status'] == 'OK':
        lat_lng = data['results'][0]['geometry']['location']
        return lat_lng['lat'], lat_lng['lng']
    else:
        return None, None
```

```python
# Loop through each address and get the latitude and longitude
results_list = []

for address in address_list:
    lat, lng = get_lat_lng(address)
    results_list.append([address, lat, lng])
```

```python
results_list
```

```python
# Convert the list of results to a DataFrame
results_df = pd.DataFrame(results_list, columns=['Address', 'Latitude', 'Longitude'])
results_df
```

```python
# export results to csv
results_df.to_csv('/Users/trdny_josephjungermann/Documents/the_real_deal/20241009_MD_Agents_Map/output/md_address_lat_lon.csv', index=False)
```
